```python
# Dependencies and Setup
import pandas as pd

# Load data
school_data = pd.read_csv("School-Data-Analysis-Pandas/Resources/schools_complete.csv")
student_data = pd.read_csv("School-Data-Analysis-Pandas/Resources/students_complete.csv")
school_data
```

```python
student_data
```

```python
# Join school_data to student_data at school_name(reason both have a school name in them)
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
school_data_complete.head(20-40)
```

```python
#calculate the total number of school name
school_count = school_data_complete["school_name"].nunique()
school_count
```

```python
# #calculate the total number of student

# student_count = len(school_data_complete["Student ID"].unique())
# student_count
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
#calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculated the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
#Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#Calculate the number of student with math scores of 70 or higher
passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
passing_math
```

```python
# percentage of student who pass math
percent_passing_math = (passing_math["Student ID"].count() / student_count * 100)
percent_passing_math
```

```python
#Calculate the number of student with math scores of 70 or higher
passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
passing_reading
```

```python
# percentage of student who pass math
percent_passing_reading = (passing_reading["Student ID"].count() / student_count * 100)
percent_passing_reading 
```

```python
#% overall passing (the percentage of students who passed math AND reading)
overall_passing_rate = (percent_passing_math + percent_passing_reading) / 2
overall_passing_rate
```

```python
# Dictrict Summary
district_summary = pd.DataFrame ({
    "Total Schools" : [school_count],
    "Total Students" : [student_count],
    "Total Budget" : [total_budget],
    "Average Math Score" : [average_math_score],
    "Average Reading Score" : [average_reading_score],
    "% Passing Math" : [percent_passing_math],
    "% Passing Reading" : [percent_passing_reading],
    "% Overall Passing" : [overall_passing_rate]
})
```

```python
district_summary
```

```python
# School Summary
# School and the type of schools
school_group = school_data_complete.groupby("school_name")
school_type = school_group["type"].first()
school_type
```

```python
# total student per school
total_student_school = school_group.size()
total_student_school
```

```python
# total budget per school
total_budget_school = school_group["budget"].first()
total_budget_school
```

```python
# total budget per student
per_student_budget = total_budget_school / total_student_school
per_student_budget
```

```python
# Average math score
average_math_score_school = school_group["math_score"].mean()
average_math_score_school
```

```python
# Average reading score
average_reading_score_school = school_group["reading_score"].mean()
average_reading_score_school
```

```python
# passing math (the percentage of students who passed math)

# passing_math_school = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()
# passing_math_school


passing_math_school = passing_math.groupby("school_name")["Student ID"].count()
passing_math_school
```

```python
# passing reading(the percentage of students who passed reading)
passing_reading_school = passing_reading.groupby("school_name")["Student ID"].count()
passing_reading_school
```

```python
#% passing math (the percentage of students who passed math)
percent_passing_math_school = (passing_math_school /total_student_school) * 100
percent_passing_math_school
```

```python
#% passing reading (the percentage of students who passed reading)
percent_passing_reading_school = (passing_reading_school /total_student_school) * 100
percent_passing_reading_school
```

```python
# % overall passing (the percentage of students who passed math AND reading)
Overall_passing_rate_school = (percent_passing_math_school + percent_passing_reading_school)/2
Overall_passing_rate_school
```

```python
#School Summary DataFrame
school_summary = pd.DataFrame ({
    "School Type" : school_type,
    "Total Students" : total_student_school,
    "Total Budget" : total_budget_school,
    "Average Math Score" : average_math_score_school,
    "Average Reading Score" : average_reading_score_school,
    "% Passing Math" : percent_passing_math_school,
    "% Passing Reading" : percent_passing_reading_school,
    "% Overall Passing" : Overall_passing_rate_school
})
school_summary
```

```python
# Highest-Performing School
top_schools = school_summary.sort_values(by="% Overall Passing", ascending=False).head()
top_schools 
```

```python
# Lowest-Performing Schools
bottom_schools = school_summary.sort_values(by="% Overall Passing").head()
bottom_schools
```

```python
# Math Scores by Grade
math_scores_by_grade = pd.pivot_table(school_data_complete, values='math_score', index='school_name', columns='grade', aggfunc='mean', fill_value=0)
math_scores_by_grade
```

```python
# Reading Scores by Grade
reading_scores_by_grade = pd.pivot_table(school_data_complete, values='reading_score', index='school_name', columns='grade', aggfunc='mean', fill_value=0)
reading_scores_by_grade
```

```python
# Scores by School Spending
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summary["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, bins=spending_bins, labels=spending_labels)
spending_summary = school_summary.groupby("Spending Ranges (Per Student)").agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    "% Passing Math": "mean",
    "% Passing Reading": "mean",
    "% Overall Passing": "mean"
})
```

```python
# Scores by School Size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary["School Size"] = pd.cut(total_student_school, bins=size_bins, labels=size_labels)
size_summary = school_summary.groupby("School Size").agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    "% Passing Math": "mean",
    "% Passing Reading": "mean",
    "% Overall Passing": "mean"
})
```

```python
# Scores by School Type
type_summary = school_summary.groupby("School Type").agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    "% Passing Math": "mean",
    "% Passing Reading": "mean",
    "% Overall Passing": "mean"
})
```

```python
# Print the DataFrames or save them to CSV as needed
print("District Summary:")
print(district_summary)

print("\nSchool Summary:")
print(school_summary)

print("\nTop Performing Schools:")
print(top_schools)

print("\nBottom Performing Schools:")
print(bottom_schools)

print("\nMath Scores by Grade:")
print(math_scores_by_grade)

print("\nReading Scores by Grade:")
print(reading_scores_by_grade)

print("\nScores by School Spending:")
print(spending_summary)

print("\nScores by School Size:")
print(size_summary)

print("\nScores by School Type:")
print(type_summary)
```

```python
# Save the outputs in an Excel file with different sheets
with pd.ExcelWriter('school_district_analysis.xlsx') as writer:
    district_summary.to_excel(writer, sheet_name='District Summary', index=False)
    school_summary.to_excel(writer, sheet_name='School Summary', index=False)
    top_schools.to_excel(writer, sheet_name='Top Performing Schools', index=False)
    bottom_schools.to_excel(writer, sheet_name='Bottom Performing Schools', index=False)
    math_scores_by_grade.to_excel(writer, sheet_name='Math Scores by Grade')
    reading_scores_by_grade.to_excel(writer, sheet_name='Reading Scores by Grade')
    spending_summary.to_excel(writer, sheet_name='Scores by School Spending')
    size_summary.to_excel(writer, sheet_name='Scores by School Size')
    type_summary.to_excel(writer, sheet_name='Scores by School Type')
```

```python
# Save the outputs in separate CSV files
district_summary.to_csv('district_summary.csv', index=False)
school_summary.to_csv('school_summary.csv', index=False)
top_schools.to_csv('top_performing_schools.csv', index=False)
bottom_schools.to_csv('bottom_performing_schools.csv', index=False)
math_scores_by_grade.to_csv('math_scores_by_grade.csv', index=True)
reading_scores_by_grade.to_csv('reading_scores_by_grade.csv', index=True)
spending_summary.to_csv('scores_by_school_spending.csv', index=True)
size_summary.to_csv('scores_by_school_size.csv', index=True)
type_summary.to_csv('scores_by_school_type.csv', index=True)
```
